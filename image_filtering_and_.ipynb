```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load an image
# Make sure you have an image file (e.g., 'my_image.jpg') in the same directory
try:
    pil_image = Image.open('my_image.jpg')
    # Convert the image to a NumPy array
    image_array = np.array(pil_image)
    print("Image loaded successfully with shape:", image_array.shape)
except FileNotFoundError:
    print("Please make sure you have an image file named 'my_image.jpg' in the same directory.")
    # Create a dummy array if the image is not found to allow the rest of the code to run
    image_array = np.random.randint(0, 256, size=(100, 150, 3), dtype=np.uint8)

plt.imshow(image_array)
plt.title("Original Image")
plt.show()
# image_array
```

```python
def to_grayscale(image_array):
    #coverts an RGB image to grayscale
    if image_array.ndim!=3 and image_array.shape[2]!=3:
        raise ValueError("Input must be a 3-channel RGB image.")

    #using luminosity method 
    r, g,b=image_array[:,:,0], image_array[:,:,1], image_array[:,:,2]
    #standard formula for gray scale conversion
    grayscale = 0.299 * r + 0.587 * g + 0.114 * b
    return grayscale.astype(np.uint8)
grayscale_image = to_grayscale(image_array)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(grayscale_image, cmap='gray')
# plt.subplot(1, 2, 2)
# plt.imshow(image_array[:,:,0], cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(image_array, cmap='gray')#ignore cmap='gray' by matplotlib for 3 dimensional image
```
