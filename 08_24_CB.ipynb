```python
#check for prime or not
n=int(input("take a number: "))
i=1
cnt=0
while(i<=n):
    if(n%i==0):
        print(i)
        cnt+=1
    i+=1
print("no of factors",cnt)
if(cnt==2):
    print("its a prime number")
else:
    print("its a composite number")
```

```python
#flag and break uses in the last statement
#check for prime or not
n=int(input("take a number: "))
i=2
c=True
while(i<n):
    if(n%i==0):
        print(i)
        c=False
        print("its a composite number")
        break
    i+=1

if(c==True):
    print("its a prime number")
```

```python
#python specific prime number code with while and else (without break and flag)
n=int(input("take a number: "))
i=2
while(i<n):
    if(n%i==0):
        print("composite no")
        break
    i+=1
else:
    print("prime no")
```

```python
#optimisation of code ->prime number
#1--> check till n/2 only 
#2--> check till sqrt of n (because after that the factors will be repeated )

n=int(input("take a number: "))
i=2
while i*i<=n:
    if n%i==0:
        print("composite no")
        break
    i+=1
else:
    print("prime number")

```
