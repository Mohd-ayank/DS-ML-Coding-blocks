```python
#pattern 13
#mirror pattern
n=5
row=1
nst=1

while row<=2*n-1:
    cst=1
    while cst<=nst:
        print("*", end=" ")
        cst+=1

    #mirror condition
    
    if row<n:
        nst+=1
    else:
        nst-=1

    row+=1
    print()
  
```

```python
#pattern 18
n=int(input("enter a number"))
row=1
nsp=n//2
nst=1

while row<=n:
    csp=1
    while csp<=nsp:
        print(" ", end=" ")
        csp+=1
    cst=1
    while cst<=nst:
        print("*", end=" ")
        cst+=1
        
    if row<=n//2:
        nsp-=1
        nst+=2
    else:
        nsp+=1
        nst-=2
        
    #write this line at the end of the code
    row+=1
    print()
```
