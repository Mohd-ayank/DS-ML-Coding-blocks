```python
#function overriding and DUNDERS -->> magic function

class Car:
    def __init__(self, model, color, speed):
        self.model=model
        self.color=color
        self.speed=speed

    # operator overloading (function overriding)
    def __eq__(self, oth):
        return self.model == oth.model

    #greater than 
    def __gt__(self, oth):
        return self.speed < oth.speed
    def __len__(self):
        return self.speed

    
    #print(c2)
    def __str__(self):
        return f"Model: {self.model}, Color:{self.color}, Speed:{self.speed}"
    
```

```python
# dir(Car)
```

```python
c1=Car("Toyota","White",30)

c2=Car("Toyota","White",30)
```

```python
c1==c2 #addresses are different (but we can modify aur dunders to modify definition of comparison)
```
