```python
pop= np.loadtxt('populations.txt')
pop
```

```python
#find max population for every year and whose population
print(np.max(pop, axis=1))
# Find the index of the maximum population in each column
max_indices_row = np.argmax(pop, axis=1)
max_indices_row
```

```python
# Extract columns using slicing
years = pop[:, 0]
populations = pop[:, 1:]
print(years)
print(populations)
```

```python
#find the years for these 3 categories maximum population
max_indices_col=np.argmax(populations, axis=0)
print(max_indices_col)
```

```python
max_years_categories=years[max_indices_col]
print(max_years_categories)
```

```python
#3rd problem- get me mean population of every year and every category
#mean population of every year
np.mean(populations, axis=1)
```

```python
#mean population of every category
np.mean(populations, axis=0)
#other way
#np.mean(pop[:,1:], axis=0) -> extracting just population columns and filtering out year col(we dont need it)
```
