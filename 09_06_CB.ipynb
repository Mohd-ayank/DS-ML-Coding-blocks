```python
#recursion
#01- return factorial of n

def fact(n):
    if(n==0 or n==1):
        return 1
    return n*fact(n-1)

fact(5)
```

```python
#lambda function returning division and modulo operation
def cal(fn,a=10,b=20):
    val=fn(a,b)
    return val

# cal(lambda a,b:a%b,a//b,103,20)#become 5 parameters

cal(lambda a,b:(a%b,a//b),103,20)#remain 4 parameters tuple-> (a//b, a%b)
```
