```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#exclude values nearer to zero otherwise cosec and tanntends to very large values and numpy might not be able to handle those
X=np.linspace(-2,2,1000)


ySin=np.sin(2*np.pi*X)
yCos=np.cos(2*np.pi*X)
yCosec=1/np.sin(2*np.pi*X)
ySec=1/np.cos(2*np.pi*X)
yCot=1/np.tan(2*np.pi*X)
yTan=np.tan(2*np.pi*X)

plt.figure(figsize=(20,10))

plt.subplot(3,2,1)
plt.plot(X,ySin)
plt.grid()
plt.title('Sin wave')

plt.subplot(3,2,2)
plt.plot(X,yCos)
plt.grid()
plt.title('Cos wave')

plt.subplot(3,2,3)
plt.ylim(-10,10)
plt.plot(X,yCosec)
plt.grid()
plt.title('Cosec wave')

plt.subplot(3,2,4)
plt.ylim(-10,10)
plt.plot(X,ySec)
plt.grid()
plt.title('Sec wave')

plt.subplot(3,2,5)
plt.ylim(-10,10)
plt.plot(X,yTan)
plt.grid()
plt.title('Tan wave')

plt.subplot(3,2,6)
plt.ylim(-10,10)
plt.plot(X,yCot)
plt.grid()
plt.title('Cot wave')

plt.subplots_adjust(hspace=0.5)
plt.show()
```
