```python
#pattern 23
n=int(input("enter a number"))
row=1
nsp=n-1
nst=1
while row<=n:
    csp=1
    while csp<=nsp:
        print(" ", end=" ")
        csp+=1

    cst=1
    while cst<=nst:
        print(1,end=" ")
        cst+=1
    nst+=2
    nsp-=1
    row+=1
    print()
```

```python
#pattern 24
#using val and cval in formal way
n=int(input("enter a number"))
row=1
nsp=n-1
nst=1
val=1
while row<=n:
    csp=1
    while csp<=nsp:
        print(" ", end=" ")
        csp+=1

    cst=1
    cval=val
    while cst<=nst:
        print(cval,end=" ")
        cst+=1
    nst+=2
    nsp-=1
    val+=1
    row+=1
    print()
```

```python
#pattern 27
n=int(input("enter a number"))
row=1
nsp=n-1
nst=1
val=1
while row<=n:
    csp=1
    while csp<=nsp:
        print(" ", end=" ")
        csp+=1

    cst=1
    cval=val
    while cst<=nst:
        print(cval,end=" ")
        if cst<=nst//2:
            cval+=1
        else:
            cval-=1
        cst+=1
    nst+=2
    nsp-=1
    row+=1
    print()
```

```python
#pattern 29
n=int(input("enter a number"))
row=1
nsp=n-1
nst=1
val=1
while row<=n:
    csp=1
    while csp<=nsp:
        print(" ", end=" ")
        csp+=1

    cst=1
    cval=val
    while cst<=nst:
        print(cval,end=" ")
        if cval==val or cval==nst:
            print(val, end=" ")
        else:
            print("0", end=" ")
        
        cst+=1
        
    nst+=2
    nsp-=1
    row+=1
    val+=1
    print()
```

```python
#pattern 32
n=int(input("enter a number"))
row=1
nst=1
val=1
while row<=2*n-1:
    cst=1
    cval=val
    while cst<=nst:
        if cst%2==0:
            print("*", end=" ")
        else:
            print(cval, end=" ")
        cst+=1
        
    if row<n:
        nst+=2
        val+=1
    else:
        nst-=2
        val-=1
    row+=1
    
    print()
```
