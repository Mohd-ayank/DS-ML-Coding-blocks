```python
#print all substrings
def substring(st,i,s):
    if i==len(st):
        return
    s+=st[i]
    print(s)
    substring(st,i+1,s)

st="Hello"
i=0
l=len(st)
while i<l:
    s=""
    substring(st,i,s)
    i+=1
```

```python
#args in functions
def operations(*args):
    if len(args)==1:
        print("only one elemet",args[0]+10)
    elif len(args)==2:
        print("only two elemets",args[0]*args[1])
    else:
        s=0
        for ele in args:
            s+=ele
        print("sum of all ele is:", s)
operations(1)
operations(2,3)
operations(10,20,30,40)
        
```

```python
#function within function

def fn1():
    x=10
    y=20
    def fn2():
        x=1000
        print(x,y)
    return fn2
    
fn2=fn1()
print(fn2())  #returning none but printing (1000,20)
```

```python
#function decorators
#take username and password from input and match it with storage in your dictionary
d={"ayan":1020, "ansh":2331, "shubham":2993}
def match(user, password):
    for k,v in d.items():
        if k==user and v==password:
            print("Approved authentication")
            break
        else:
            print("request revoked")
            break

user=input()
password=int(input())
match(user, password)
```
